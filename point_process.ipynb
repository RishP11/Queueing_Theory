```python
import numpy as np 

import matplotlib.pyplot as plt 
%config InlineBackend.figure_format='svg'
import seaborn as sns 

sns.set_theme('paper')
```

```python
# Stationary Poisson Point process:
xlims = [-10, 10]
ylims = [-10, 10]

lmbda = 2
A = (xlims[1] - xlims[0]) * (ylims[1] - ylims[0])
num_points = np.random.poisson(lmbda * A)
xs = np.random.uniform(xlims[0], xlims[1], num_points)
ys = np.random.uniform(ylims[0], ylims[1], num_points)

plt.scatter(xs, ys, color='purple')
```
